```python
def imshow(title="",image=None,size=3):
  h,w=image.shape[0],image.shape[1]
  aspect_ratio=w/h


  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path='/content/drive/MyDrive/MyDataSet/Class1/nature2.jpg'
image1=cv2.imread(image_path)
```

```python
B, G ,R = cv2.split(image1)

print(B.shape)
print(G.shape)
print(R.shape)
```

```python
imshow("onyl blue channel",B,10)
```

```python
zeros=np.zeros(image1.shape[:2],dtype="uint8")
```

```python
print(zeros.shape)
print(R.shape)
print(G.shape)
print(B.shape)
```

```python
imshow("Red", cv2.merge([zeros, zeros, R]))
imshow("Green", cv2.merge([zeros, G, zeros]))
imshow("Blue", cv2.merge([B, zeros, zeros]))
```

```python
imshow("Combined", cv2.merge([B, G, R]))
```
